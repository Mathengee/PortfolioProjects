```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\OneDrive\Documents\Pandas Tutorial\Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor').mean(['Flavor Rating','Texture Rating','Total Rating'])
```

```python
group_by_frame.mean()
```

```python
# Select only numeric columns and calculate their mean by 'Base Flavor'
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby('Base Flavor')[numeric_cols].mean()
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min()
```
